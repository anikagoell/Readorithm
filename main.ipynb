Model Building

```python
rating_name=rating.merge(book,on='ISBN')
numRatingBook=rating_name.groupby('Title').count()['User-ID'].reset_index()
numRatingBook.rename(columns={'User-ID':'Rating Count'},inplace=True)
numRatingBook
```

```python
avgRatingBook=rating_name.groupby('Title')['Book-Rating'].mean().reset_index()
avgRatingBook.rename(columns={'Book-Rating':'Rating Avg'},inplace=True)
avgRatingBook
```

```python
ratingNew = numRatingBook.merge(avgRatingBook, on=('Title')) #60 popular books based on ratings and also popularity based recommendation
ratingNew=ratingNew[ratingNew['Rating Count']>249].sort_values('Rating Avg',ascending=False).head(60)
ratingNew
```

```python
pop_data=ratingNew.merge(book,on=('Title')).drop_duplicates('Title')
pop_data=pop_data[['Title','Rating Count','Rating Avg','ISBN','Author','Published-Year','URL-M']].reset_index(drop=True)
pop_data
```

Collaborative approach


-only users who have rated more than 230 books
-only books which have recieved atleast 50 votes

```python
userRatedBooks=rating_name.groupby('User-ID').count()
top230users=userRatedBooks.loc[userRatedBooks['Book-Rating']>=230].index
top230users
```

```python
filterRating= rating_name[rating_name['User-ID'].isin(top230users)]
filterRating
```

```python
y=filterRating.groupby('Title').count()['Book-Rating']>=50
famousBooks=y[y]
```

```python
famousBooks=famousBooks.index
```

```python
finalModel=filterRating[filterRating['Title'].isin(famousBooks)]
ptable= finalModel.pivot_table(index='Title', columns='User-ID', values='Book-Rating')
```

```python
ptable.fillna(0, inplace=True)
```

```python
ptable
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cos
```

```python
cosTable=cos(ptable)
```

```python
cos(ptable).shape
```

```python
cosTable
```

```python
def recom(bookName):
    index = np.where(ptable.index==bookName)[0][0]
    similiar= sorted(list(enumerate(cosTable[index])),key=lambda x:x[1],reverse=True)[1:8]
    data=[]
    for i in similiar:
        item=[]
        temp=book[book["Title"]==ptable.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Title')['Title'].values))
        item.extend(list(temp.drop_duplicates('Title')['Author'].values))
        item.extend(list(temp.drop_duplicates('Title')['URL-L'].values))
        item.extend(list(temp.drop_duplicates('Title')['Published-Year'].values))
        data.append(item)
    return data
```

```python
recom('Message in a Bottle')
```
